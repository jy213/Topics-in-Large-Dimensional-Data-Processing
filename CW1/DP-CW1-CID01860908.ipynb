```julia
function Find_Convolution_Matrix(x, h)
    n = length(x)
    m = length(h)
    v_zeros = zeros(n-1)
    h_new = vcat(v_zeros, vcat(h, v_zeros))
    C = zeros((n+m-1), n)
    for i in 1:(n+m-1)
        C[i, :] = reverse(h_new[i:i+n-1])
    end
    return C
end

function Find_Convolution_Adjoint_Matrix(x, h)
    C = Find_Convolution_Matrix(x, h)
    return adjoint(C)
end
        
CW1_DATA.Q4_3_3_a_ans_Operator = Find_Convolution_Matrix(CW1_DATA.Q4_3_3_data_x,CW1_DATA.Q4_3_3_data_h)
CW1_DATA.Q4_3_3_a_ans_AdjointOperator = Find_Convolution_Adjoint_Matrix(CW1_DATA.Q4_3_3_data_x,CW1_DATA.Q4_3_3_data_h)
```

### 5.(a)

First considering about the 2D Discrete Fourier Transform, the followings are the equations for 1D Discrete Fourier Transform and 2D Discrete Fourier Transform:
$$
\begin{equation*}
\begin{split}
    \bm{F[l]} &= \bm{\sum _{n=0}^{N-1}f[n]e^{-j2\pi\frac{l}{N}n}} \\
    \bm{F[k, l]} &= \bm{\sum _{m=0}^{M-1}\sum _{n=0}^{N-1}f[m, n]e^{-j2\pi\left(\frac{k}{M}m+\frac{l}{N}n\right)}}
\end{split}
\end{equation*}
$$
Now try to find the relationship between 2D Discrete Fourier Transform and 1D Discrete Fourier Transform so that we could use fft() to do the 2D DFT:
$$
\begin{equation*}
    \bm{F[k, l]} = \bm{\sum _{m=0}^{M-1}\sum _{n=0}^{N-1}f[m, n]e^{-j2\pi\left(\frac{k}{M}m+\frac{l}{N}n\right)}} = \bm{\sum _{m=0}^{M-1}\sum _{n=0}^{N-1}f[m, n]e^{-j2\pi\frac{k}{M}m}e^{-j2\pi\frac{l}{N}n}} = \bm{\sum _{m=0}^{M-1}\left(\sum _{n=0}^{N-1}f[m, n]e^{-j2\pi\frac{l}{N}n}\right)e^{-j2\pi\frac{k}{M}m}}
\end{equation*}
$$
It can be seen from the above equation that the 2D DFT can be separated into performing a 1D DFT for each row/column vector of a matrix first, and then a 1D DFT for each column/row vector of the modified matrix.

Now considering about the fftshift operator, the 2D fftshift can also be explained by first doing the 1D fftshift for each row/column vector of a matrix, and the 1D fftshift for each column/row vector of the modified matrix. And the $\bm{P_{\Omega}}$ operation is the same as the previous one.

```julia
function Find_POmega_withoutmatrix(Omega, X)
    m, n = size(Omega)
    X_new = []
    for j in 1:n
        for i in 1:m
            X_new = (Omega[i, j] == 1) ? vcat(X_new, X[i, j]) : X_new
        end
    end
    return X_new
end

function Find_MRI_operation(Omega, X)
    m, n = size(X)
    ##### Find O1 #######
    # First compute the 1D DFT for column vectors of X
    O_1_pre = zeros(ComplexF64, m, n)
    for i in 1:n
        O_1_pre[:, i] = FFTW.fft(X[:, i])
    end
    # Then compute the 1D DFT for row vectors of O_1_pre
    O_1 = zeros(ComplexF64, m, n)
    for i in 1:n
        O_1[i, :] = FFTW.fft(O_1_pre[i, :])
    end
    ##### Find O2 #######
    # First compute the 1D fftshift for row vectors of O1
    O_2_pre = zeros(ComplexF64, m, n)
    for i in 1:m
        O_2_pre[i, :] = FFTW.fftshift(O_1[i, :])
    end
    # Then compute the 1D fftshift for column vectors of O_2_pre
    O_2 = zeros(ComplexF64, m, n)
    for i in 1:n
        O_2[:, i] = FFTW.fftshift(O_2_pre[:, i])
    end
    ##### Find O3 #######
    O_3 = Find_POmega_withoutmatrix(Omega, O_2)
    return O_1, O_2, O_3
end

O_1, O_2, O_3 = Find_MRI_operation(CW1_DATA.Q4_3_5_data_Omega, CW1_DATA.Q4_3_5_data_X)
CW1_DATA.Q4_3_5_a_ans_O1 = O_1
CW1_DATA.Q4_3_5_a_ans_O2 = O_2
CW1_DATA.Q4_3_5_a_ans_O3 = O_3
println("O_1: ", CW1_DATA.Q4_3_5_a_ans_O1)
println("O_2: ", CW1_DATA.Q4_3_5_a_ans_O2)
println("O_3: ", CW1_DATA.Q4_3_5_a_ans_O3)
```

### 5.(b)

Considering about the adjoint matrix of 2D DFT operator and 2D fftshift operator, we first look at the 2D DFT since we know that the inverse of 1D DFT is
$$
\begin{equation*}
    \bm{f[l]} = \bm{\frac{1}{N}}\bm{\sum _{n=0}^{N-1}F[n]e^{j2\pi\frac{l}{N}n}}
\end{equation*}
$$
It can be seen that the form is exactly same as the DFT except there's a $\bm{\frac{1}{N}}$ scaling and the minus sign becomes positive sign, which is exactly the conjugate of the original function. Furthermore since the linear operator of DFT is symmetric, then the transpose of this operator is equal to itself, so we could say that the conjugate transpose of DFT operator is equal to the inverse DFT operator with a scaling difference. Additionally, since 2D DFT is just a combination of two 1D DFT, then the adjoint operation of 2D DFT is equal to the combination of the corresponding 1D inverse DFT operations with extra care of the scaling. 

For fftshift, if we consider about the general case, where the dimension of $\bm{X}$ is $\bm{m}\times \bm{n}$, then the output of this operation should also be a matrix with the same dimension, then it's impossible to have a linear operator that can directly multiply with $\bm{X}$ to get the output, we have to first vectorize the matrix $\bm{X}$ and the output would also be the vectorization form. In this case, since the input and output vector has the same dimension, and the basic logic of this operation is to change the position of each element of input vector, then the linear operator would have the form that there is only one 1 in each column with other elements be 0, and the position of 1 will be different for each column vector, since this operator is also a symmetric matrix, it's actually an orthonormal matrix, since all the elements are 1 and 0 which is real, then its conjugate transpose is actually equal to its inverse, which means the adjoint operation of fftshift is actually its inverse. The inverse of fftshift is actually the fftshift of the shifted vector if its length is even, and when the dimension is odd, then we add a 0 into the middle point and do the fftshift for the shifted vector and then get rid of the last 0.

```julia
function Adjoint_Subsampling(Omega, y)
    m, n = size(Omega)
    Z = zeros(ComplexF64, m, n)
    index = 1
    for j in 1:n
        for i in 1:m
            Z[i, j] = (Omega[i, j] == 0) ? 0 : y[index]
            index = (Omega[i, j] == 0) ? index : index+1
        end
    end
    return Z
end

function Find_Adjoint_MRI_withoutmatrix(Omega, y)
    m, n = size(Omega)
    ####### Find O1 ########
    # Find the O1 using the adjoint function of subsampling operation
    O_1 = Adjoint_Subsampling(Omega, y)
    ####### Find O2 ########
    # First find the adjoint operation of 1D fftshift for comlumn vectors of O1
    # if the m is even, do the inverse fftshift as normal, if is odd, add a 0 and then do fftshift then get rid of 0
    O_2_pre = zeros(ComplexF64, m, n)
    if (m%2 == 1)
        mid = Int((m-1)/2)
        for i in 1:n
            O_2_pre[:, i] = FFTW.fftshift(vcat(O_1[1:mid, i], vcat(0, O_1[(mid+1):m, i])))[1:m]
        end
    else
        for i in 1:n
            O_2_pre[:, i] = FFTW.fftshift(O_1[:, i])
        end
    end
    # similarly for row vectors
    O_2 = zeros(ComplexF64, m, n)
    if (n%2 == 1)
        mid = Int((n-1)/2)
        for i in 1:m
            O_2[i, :] = FFTW.fftshift(vcat(O_2_pre[i, 1:mid], vcat(0, O_2_pre[i, (mid+1):n])))[1:n]
        end
    else
        for i in 1:m
            O_2[i, :] = FFTW.fftshift(O_2_pre[i, :])
        end
    end
    ####### Find O3 ########
    O_3_pre = zeros(ComplexF64, m, n)
    # Find the inverse fft for row vectors of O2
    for i in 1:n
        O_3_pre[i, :] = FFTW.ifft(O_2[i, :])
    end
    # Find the inverse fft for column vectors of O_3_pre
    O_3 = zeros(ComplexF64, m, n)
    for i in 1:m
        O_3[:, i] = FFTW.ifft(O_3_pre[:, i])
    end
    O_3 = O_3*m*n
    return O_1, O_2, O_3 
end

O1, O2, O3 = Find_Adjoint_MRI_withoutmatrix(CW1_DATA.Q4_3_5_data_Omega, CW1_DATA.Q4_3_5_a_ans_O3)
CW1_DATA.Q4_3_5_b_ans_O1 = O1
CW1_DATA.Q4_3_5_b_ans_O2 = O2
CW1_DATA.Q4_3_5_b_ans_O3 = O3
println("O_1: ", CW1_DATA.Q4_3_5_b_ans_O1)
println("O_2: ", CW1_DATA.Q4_3_5_b_ans_O2)
println("O_3: ", CW1_DATA.Q4_3_5_b_ans_O3)
```

## 5.1. The Netflix Problem

### 1.(a)

Original Equation:
$$
\begin{equation*}
	\begin{split}
		V^{k+1} &= \arg \min_{\bm{V}}~ \lVert \bm{y} - \mathcal{P}_{\Omega}(\bm{U_k} \bm{V}^{\mathsf{T}}) \rVert_2^2 \\ 
		U^{k+1} &= \arg \min_{\bm{U}}~ \lVert \bm{y} - \mathcal{P}_{\Omega}(\bm{U} \bm{V}^{k+1, \mathsf{T}}) \rVert_2^2 \\ 
	\end{split}
\end{equation*}
$$
It can be seen that this is a standard form of least square, the only difference here is that the input to $\bm{P_{\Omega}}$ should be first vectorized. 

First considering about the updates for $\bm{V}$, assume that $\bm{vect(U_kV) = x}$, $\bm{P_{\Omega}(U_kV^{\mathsf{T}})}$ is equivalent to an operator multiplying the vectorization of $\bm{U_kV^{\mathsf{T}}}$, which is $\bm{P_{\Omega}x}$

Therefore, we could rewrite the above least square as 
$$
\begin{equation*}
	\bm{f(x)} = \| \bm{y - P_{\Omega}x} \|^2
\end{equation*}
$$
And set 
$$
\begin{equation*}
	\nabla \bm{f(x)} = \bm{P_{\Omega}}^{\mathsf{T}}(\bm{P_{\Omega}}\bm{x}-\bm{y})
\end{equation*}
$$
let $\nabla \bm{f(x)} = 0$, then 
$$
\begin{equation*}
    (\bm{P_{\Omega}}^{\mathsf{T}}\bm{P_{\Omega}})\hat{\bm{x}} = \bm{P_{\Omega}}^{\mathsf{T}}\bm{y}
\end{equation*}
$$
Since $\bm{P_{\Omega}}$ has a very large dimension, and it would be very inefficient to calculate operator $\bm{P_{\Omega}}$ from $\bm{\Omega}$ because we would have two for loops which have time complexity of around $\bm{O(MN)}$. Therefore, it's important to find a way to get the answer without calculating this matrix. Now looking at this equation, we could directly compute the $\bm{P_{\Omega}}^{\mathsf{T}}\bm{y}$ without generating the $\bm{P_{\Omega}}$ operator because it's just the adjoint subsampling operation of $\bm{y}$. Now considering the left side of the equation, we know that $\bm{P_{\Omega}}$ is in the form that all rows have 1 at only one index with all other elements be 0, and these rows are independent. Based on this form, $\bm{P_{\Omega}^{\mathsf{T}}P_{\Omega}}$ is actually the combination of $\bm{P_{\Omega}}$ and some purely zeros raw vectors. So when it multiplies with vectorized $\bm{x}$, the value of element of vectorized x will be maintined if the corresponding index of $\Omega$ is 1, if the element of $\Omega$ is zero, then it would also be zero. The adjoint subsampling operation to $\bm{y}$ is also to add the lossing zeros to the corresponding position based on $\bm{\Omega}$. Therefore, $\bm{P_{\Omega}^{\mathsf{T}}P_{\Omega}x}$ would be in the form of 
$$
\begin{bmatrix}
    \bm{X} \\
    \bm{X} \\
	\vdots \\
    0 \\
    \vdots \\
    \bm{X}\\
	0\\
	\vdots\\
	\bm{X}
\end{bmatrix}
$$
$\bm{P_{\Omega}^{\mathsf{T}}y}$ would have exactly the same form with zeros at the same positions, therefore to make two sides equal, we could set all non-zero terms of $\bm{x}$ and $\bm{y}$ to be equal, and we have the freedom to choose the values of other elements of $\bm{x}$. Since we could always decompose input matrix to $\bm{U}$ and $\bm{V}$, and the rank would be determined by the lowest rank of $\bm{U}$ and $\bm{V}$ with one dimension equal to 10, the rank requirement is always satisfied and we don't need to consider about it. Therefore, without loss of generalisity, we choose those free variables of $\bm{x}$ to be 0, then in this case, 
$$
\begin{equation*}
	(\bm{P_{\Omega}}^{\mathsf{T}}\bm{P_{\Omega}})\hat{\bm{x}} = \bm{\hat{x}} = \bm{P_{\Omega}}^{\mathsf{T}}\bm{y}
\end{equation*}
$$
Therefore 
$$
\begin{equation*}
	\bm{UV^{\mathsf{T}}} = \bm{devect\left(P_{\Omega}^{\mathsf{T}}y\right)}
\end{equation*}
$$
where devect is simply the reverse of vectorization.
Therefore, if we want to update $\bm{V}$, we could write that as 
$$
\begin{equation*}
	\bm{V_{k+1}^{\mathsf{T}}} = \bm{U_k^{\dagger}U_kV_{k+1}} = \bm{U_k^{\dagger}P_{\Omega}^{\ast}(y)}
\end{equation*}
$$
Then
$$
\begin{equation*}
	\bm{V_{k+1}} = \bm{\left(U_k^{\dagger}P_{\Omega}^{\ast}(y)\right)^{\mathsf{T}}}
\end{equation*}
$$
Similarly
$$
\begin{equation*}
	\bm{U_{k+1}} = \bm{P_{\Omega}^{\ast}(y)\left(V_{k+1}^{\mathsf{T}}\right)^{\dagger}}
\end{equation*}
$$


```julia
function Adjoint_Subsampling(Omega, y)
    m, n = size(Omega)
    Z = zeros(m, n)
    index = 1
    for j in 1:n
        for i in 1:m
            Z[i, j] = (Omega[i, j] == 0) ? 0 : y[index]
            index = (Omega[i, j] == 0) ? index : index+1
        end
    end
    return Z
end

function Adjoint_Subsampling_vector(Omega, y)
    m, n = size(Omega)
    v = zeros(m*n)
    index = 1
    for j in 1:n
        for i in 1:m
            if Omega[i, j] == 1
                v[(j-1)*m+i] = y[index]
                index += 1
            end
        end
    end
    return v
end

# Using this way to do the subsampling can increase the speed a lot by getting rid of two big for loops
function SubSampling_Operation(Omega, X)
    m, n = size(Omega)
    X_new = zeros(m*n)
    X_dot = Omega.*X
    for i in 1:n
        X_new[(i-1)*m+1:i*m] = X_dot[1:end, i] 
    end
    return X_new
end  
Omega = [0 1 0; 0 1 1; 1 0 0; 1 1 0]
y = [1, 2, 3, 4, 5, 6] 
v = Adjoint_Subsampling_vector(Omega, y)   
```

```julia
function Update_V(Omega, y, U_k)
    P_y_adjoint = Adjoint_Subsampling(Omega, y)
    U_pin = pinv(U_k)
    return transpose(U_pin*P_y_adjoint)
end

V_1 = Update_V(CW1_DATA.Q5_1_data_Omega, CW1_DATA.Q5_1_data_y, CW1_DATA.Q5_1_data_Uini) 
CW1_DATA.Q5_1_a_ans_V = V_1
```

```julia
function Update_U(Omega, y, V_kplus1)
    P_y_adjoint = Adjoint_Subsampling(Omega, y)
    V_kplus1_T_pinv = pinv(transpose(V_kplus1))
    return P_y_adjoint*V_kplus1_T_pinv     
end

U_1 = Update_U(CW1_DATA.Q5_1_data_Omega, CW1_DATA.Q5_1_data_y, CW1_DATA.Q5_1_a_ans_V)
CW1_DATA.Q5_1_b_ans_U = U_1
```

```julia
function Alternating_Minimization(Omega, y, U_1, V_1)
    y_est = SubSampling_Operation(Omega, U_1*transpose(V_1))
    y_subsamp = Adjoint_Subsampling_vector(Omega, y)
    y_error_cur = norm(y_subsamp - y_est)
    Error_dif = Inf
    y_error_pre = Inf
    U_next = U_1
    V_next = V_1
    iteration = 1
    while (abs(Error_dif) >= 0.000000000001)
        V_next = Update_V(Omega, y, U_next)
        U_next = Update_U(Omega, y, V_next)
        y_est = SubSampling_Operation(Omega, U_next*transpose(V_next))
        y_error_pre = y_error_cur
        y_error_cur = norm(y_subsamp - y_est)
        Error_dif = y_error_pre - y_error_cur
        iteration += 1
    end
    return U_next, V_next, y_error_cur, iteration, Error_dif
end

println("Updating the U and V:")
U, V, error, iteration, Error_dif = Alternating_Minimization(CW1_DATA.Q5_1_data_Omega, CW1_DATA.Q5_1_data_y, CW1_DATA.Q5_1_b_ans_U, CW1_DATA.Q5_1_a_ans_V)
CW1_DATA.Q5_1_c_ans_U = U
CW1_DATA.Q5_1_c_ans_V = V
println(iteration)
println(Error_dif)
    
```

## 5.2. Estimation Accuracy

Given 
$$
\begin{equation*}
    \boldsymbol{y} =  \boldsymbol{A}  \boldsymbol{x}_0 +  \boldsymbol{w}
\end{equation*}
$$
and 
\begin{equation*}
    \hat{\bm{x}} = \arg \min_{\bm{x}}~ \frac{1}{2}\lVert \bm{y} - \bm{A}\bm{x} \rVert_2^2
\end{equation*}

First we try to solve this minimum problem to get the representation of $\hat{\bm{x}}$. This is a standard form of Least Square Estimation problem, and the closed form representation can be found like this:
\begin{equation*}
    \begin{split}
        \bm{f(x)} &= \lVert \bm{y} - \bm{A}\bm{x} \rVert_2^2 \\
        \nabla \bm{f(x)} &= \bm{A}^{\mathsf{T}}(\bm{A}\bm{x}-\bm{y})\\
    \end{split}
\end{equation*}
let $\nabla \bm{f(x)} = 0$, then 
$$
\begin{equation*}
    (\bm{A}^{\mathsf{T}}\bm{A})\hat{\bm{x}} = \bm{A}^{\mathsf{T}}\bm{y}
\end{equation*}
$$
Condsidering a general case, then 
$$
\begin{equation*}
    \hat{\bm{x}} = (\bm{A}^{\mathsf{T}}\bm{A})^{\dagger}\bm{A}^{\mathsf{T}}\bm{y} = \bm{A}^{\dagger}\bm{y}
\end{equation*}
$$
Now, assume $A = \bm{U}\bm{\Sigma}\bm{V}^{\mathsf{T}}$, and $\bm{A}^{\dagger}$ is found by Singular Value Decomposition of $\bm{A}$, then $\bm{A}^{\dagger} = \bm{V}\bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}$, so 
$$
\begin{equation*}
    \hat{\bm{x}} = \bm{V}\bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{y}
\end{equation*}
$$
This is the closed form of the first minimum requirement, we could substitute this expression into $ \| \hat{\bm{x}} - \bm{x_0} \|^2$, then it becomes
$$
\begin{equation*}
    \| \bm{V}\bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{y} - \bm{x_0} \|^2 = \| \bm{V}\bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}(\bm{A}\bm{x_0}+\bm{w}) - \bm{x_0} \|^2
\end{equation*}
$$
Expand this expression, we could get
$$
\begin{equation*}
    \| \bm{V}\bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{A}\bm{x_0} + \bm{V}\bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{w} - \bm{x_0} \|^2 = \| \bm{V}\bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{U}\bm{\Sigma}\bm{V}^{\mathsf{T}}\bm{x_0} + \bm{V}\bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{w} - \bm{x_0} \|^2
\end{equation*}
$$
Since $\bm{A}$ is a square matrix, and $\bm{V}\bm{V}^{\mathsf{T}} = \bm{I}$, and $\bm{U}^{\mathsf{T}}\bm{U} = \bm{I}$, then 
$$
\begin{equation*}
    \| \bm{V}\bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{U}\bm{\Sigma}\bm{V}^{\mathsf{T}}\bm{x_0} + \bm{V}\bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{w} - \bm{x_0} \|^2 = \| \bm{x_0} + \bm{V}\bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{w} - \bm{x_0} \|^2 = \| \bm{V}\bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{w} \|^2
\end{equation*}
$$

No matter for the first or second question, we could assume it's an optimization question, and the above function is the objective function of this question with constraint $\|\bm{w}\|_2 = 1$, but this constraint is not an affine function, it's quite hard to use the normal Lagrange method to solve it, considering about finding both maximum and minimum, it's better to try to find both lower bound and upper bound for this objective function to find the answer. Based on the form of objective function, we could try to use the properties of $\bm{U}, \bm{V}, \bm{\Sigma}$ matrices combined with $\|\bm{w}\|_2 = 1$ to find the maximum and minimum.

Starting from the left most, we could write 
$$
\begin{equation*}
    \| \bm{V}\bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{w} \|^2 = \| \bm{V}(\bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{w}) \|^2 = \| \bm{V}\bm{x} \|^2
\end{equation*}
$$
where $\bm{x} = \bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{w}$, 
Since 
$$
\begin{equation*}
    \| \bm{V}\bm{x} \|^2 = (\bm{V}\bm{x})^{\mathsf{T}}\bm{V}\bm{x} = \bm{x}^{\mathsf{T}}\bm{V}^{\mathsf{T}}\bm{V}\bm{x}
\end{equation*}
$$
Since $\bm{V}$ is an orthonormal matrix, then $\bm{V}^{\mathsf{T}}\bm{V} = \bm{I}$, therefore,
$$
\begin{equation*}
    \bm{x}^{\mathsf{T}}\bm{V}^{\mathsf{T}}\bm{V}\bm{x} = \bm{x}^{\mathsf{T}}\bm{I}\bm{x} = \bm{x}^{\mathsf{T}}\bm{x} = \| \bm{x} \|^2
\end{equation*}
$$
So based on above observation, we could rewrite that
$$
\begin{equation*}
    \| \bm{V}(\bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{w}) \|^2 = \| \bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{w} \|^2
\end{equation*}
$$
Now using the same logic, considering about $\| \bm{\Sigma}^{-1}(\bm{U}^{\mathsf{T}}\bm{w}) \|^2 = \| \bm{\Sigma}^{-1}(\bm{x}) \|^2$, where $\bm{x} = \bm{U}^{\mathsf{T}}\bm{w}$,
Since in our case $\bm{A}$ is a square matrix, then $\bm{\Sigma}$ should also be a square diagonal matrix, which can be represented as,
$$
\begin{equation*}
    \bm{\Sigma} = 
    \begin{bmatrix}
        \bm{\lambda _1} & 0 & 0 & 0 & \cdots & 0 \\
        0 & \bm{\lambda _2} & 0 & 0 & \cdots & 0 \\
        0 & 0 & \bm{\lambda _3} & 0 & \cdots & 0 \\
        \vdots & & & \ddots & & \vdots\\
        0 & 0 & 0 & 0 & \cdots & \bm{\lambda _n}

    \end{bmatrix} \rightarrow \bm{\Sigma}^{-1} = 
    \begin{bmatrix}
        \frac{1}{\bm{\lambda _1}} & 0 & 0 & 0 & \cdots & 0 \\
        0 & \frac{1}{\bm{\lambda _2}} & 0 & 0 & \cdots & 0 \\
        0 & 0 & \frac{1}{\bm{\lambda _3}} & 0 & \cdots & 0 \\
        \vdots & & & \ddots & & \vdots\\
        0 & 0 & 0 & 0 & \cdots & \frac{1}{\bm{\lambda _n}}

    \end{bmatrix}
\end{equation*}
$$

Therefore, we could write $\bm{\Sigma}^{-1}(\bm{x})$ as 
$$
\begin{equation*}
    \begin{bmatrix}
        \frac{1}{\bm{\lambda _1}} & 0 & 0 & 0 & \cdots & 0 \\
        0 & \frac{1}{\bm{\lambda _2}} & 0 & 0 & \cdots & 0 \\
        0 & 0 & \frac{1}{\bm{\lambda _3}} & 0 & \cdots & 0 \\
        \vdots & & & \ddots & & \vdots\\
        0 & 0 & 0 & 0 & \cdots & \frac{1}{\bm{\lambda _n}}
    \end{bmatrix}
    \begin{bmatrix}
    \bm{x_1} \\
    \bm{x_2} \\
    \bm{x_3} \\
    \vdots \\
    \bm{x_n}
    \end{bmatrix} = 
    \begin{bmatrix}
    \frac{\bm{x_1}}{\lambda _1} \\
    \frac{\bm{x_2}}{\lambda _2} \\
    \frac{\bm{x_3}}{\lambda _3} \\
    \vdots \\
    \frac{\bm{x_n}}{\lambda _n}
    \end{bmatrix}
\end{equation*}
$$
Therefore, $\| \bm{\Sigma}^{-1}(\bm{x}) \|^2$ can be written as 
$$
\begin{equation*}
\sum_{i = 1}^{n} \frac{\bm{x_i}^2}{\lambda _i^2}
\end{equation*}
$$
In our case, $\lambda _i$ is fixed and $\bm{x_i}$ is variant, but now matter how we choose $\bm{x_i}s$ we could always make sure that 
$$
\begin{equation*}
    \frac{\sum_{i = 1}^{n}\bm{x_i^2}}{\lambda _{max}^2} < \sum_{i = 1}^{n} \frac{\bm{x_i}^2}{\lambda _i^2} < \frac{\sum_{i = 1}^{n}\bm{x_i^2}}{\lambda _{min}^2}
\end{equation*}
$$
Now considering about $\sum_{i = 1}^{n}\bm{x_i^2}$, which is equal to $\| \bm{x} \|^2$, where $\bm{x} = \bm{U}^{\mathsf{T}}\bm{w}$, so 
$$
\begin{equation*}
    \| \bm{x} \|^2 = \| \bm{U}^{\mathsf{T}}\bm{w} \|^2 = (\bm{U}^{\mathsf{T}}\bm{w})^{\mathsf{T}}(\bm{U}^{\mathsf{T}}\bm{w}) = \bm{w}^{\mathsf{T}}\bm{U}\bm{U}^{\mathsf{T}}\bm{w}
\end{equation*}
$$
Since $\bm{U}$ is an orthonormal matrix, then $\bm{U}\bm{U}^{\mathsf{T}} = \bm{I}$, then
$$
\begin{equation*}
    \bm{w}^{\mathsf{T}}\bm{U}\bm{U}^{\mathsf{T}}\bm{w} = \bm{w}^{\mathsf{T}}\bm{w} = \| \bm{w} \|^2
\end{equation*}
$$
Since in our case, we have the requirement that $\| \bm{w} \|^2 = 1$, then we could know that
$$
\begin{equation*}
    \| \bm{x} \|^2 = \| \bm{U}^{\mathsf{T}}\bm{w} \|^2 = \| \bm{w} \|^2 = 1
\end{equation*}
$$
Therefore
$$
\begin{equation*}
    \frac{1}{\lambda _{max}^2} < \sum_{i = 1}^{n} \frac{\bm{x_i}^2}{\lambda _i^2} < \frac{1}{\lambda _{min}^2}
\end{equation*}
$$
This condition is always true in our case, so to minimize or maximize the $\| \hat{\bm{x}} - \bm{x_0} \|^2$, we could try to make the above expression equal to its lower bound and its upper bound respectively.

Therefore, for minimum, we could make $\sum_{i = 1}^{n} \frac{\bm{x_i}^2}{\lambda _i^2} = \frac{1}{\lambda _{max}^2}$. Since $\bm{\lambda _i}$ is fixed coming from given $\bm{A}$, we could only change $\bm{x_i}$, one case to satisfy this is when all $\bm{x_i} = 0$ except for $\bm{x_j} = 1$ where $\bm{j}$ is the index of $\bm{\lambda _j} s.t. \bm{\lambda _j} = \bm{\lambda _{max}}$. Therefore, the algorithm to find $\bm{w}$ to minimize $\| \hat{\bm{x}} - \bm{x_0} \|$ can be summarised as:
1. Finding the SVD decomposition of $\bm{A}$ to get $\bm{\Sigma}$
2. Sort the $\bm{n}$ diagonal elements(eigen values of $\bm{A}$) to find the index of the $\bm{\lambda _{max}}$
3. Set $\bm{U}^{\mathsf{T}}\bm{w} = [0, 0, \cdots, 1, 0, \cdots, 0]^{\mathsf{T}} = \bm{z}$, where 1 is at the index found from last step
4. Solving $\bm{U}^{\mathsf{T}}\bm{w} = \bm{z}$, since $\bm{U}$ in our case always has the inverse which is $\bm{U}^{\mathsf{T}}$, then $\bm{w} = \bm{U}\bm{z}$

For maximum problem, the basic procedure is the same, but in step 2, the index should be found for $\lambda _{min}$

After $\bm{w}$ is found, the error can be calculated as 
$$
\begin{equation*}
    \bm{error} = (\bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{w})^{\mathsf{T}}\bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{w}
\end{equation*}
$$

if $\bm{w}$ is calculated for minimum and maximum error, then the corresponding errors are
$$
\begin{equation*}
\begin{split}
    \bm{error_{min}} &= \frac{1}{\lambda _{max}^2} \\
    \bm{error_{max}} &= \frac{1}{\lambda _{min}^2}
\end{split}
\end{equation*}
$$


### 1.(a)

```julia
using LinearAlgebra
function Find_min_error_w(A)
    F = svd(A)
    U = F.U
    Σ = F.S
    m, n = size(A)
    max_index = 1
    max = Σ[1]
    for i in 2:n
        if Σ[i] > max
            max_index = i
            max = Σ[i]
        end
    end
    z = zeros(n)
    z[max_index] = 1
    w = U*z
    error = 1/(Σ[max_index]*Σ[max_index])
    return w, error
end

w, error = Find_min_error_w(CW1_DATA.Q5_2_data_A)
CW1_DATA.Q5_2_a_ans_w = w
CW1_DATA.Q5_2_a_ans_err = error
```

### 1.(b)

```julia
function Find_max_error_w(A)
    F = svd(A)
    U = F.U
    Σ = F.S
    m, n = size(A)
    min_index = 1
    min = Σ[1]
    for i in 2:n
        if Σ[i] < min
            min_index = i
            min = Σ[i]
        end
    end
    z = zeros(n)
    z[min_index] = 1
    w = U*z
    error = 1/(Σ[min_index]*Σ[min_index])
    return w, error
end

w, error = Find_max_error_w(CW1_DATA.Q5_2_data_A)
CW1_DATA.Q5_2_b_ans_w = w
CW1_DATA.Q5_2_b_ans_err = error
```

### 1.(c)

From the previous derivation, we could say that
$$
\begin{equation*}
    \| \hat{\bm{x}} - \bm{x_0} \|^2 = \| \bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{w} \|^2
\end{equation*}
$$
Since $\bm{w}$ has i.i.d Gaussian entries, which means all elements of $\bm{w}$ are independent random variables with the same probability, which is $\bm{w_i} \sim \bm{N(\mu, \sigma)}$. Our goal is to find the expectation of error, which can be represented as
$$
\begin{equation*}
    \bm{E[ \| \hat{\bm{x}} - \bm{x_0} \|^2 ]} = \bm{E[\| \bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{w} \|^2]}
\end{equation*}
$$
To expand it
$$
\begin{equation*}
    \bm{E[\| \bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{w} \|^2]} = \bm{E[(\bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{w})^{\mathsf{T}}(\bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{w})]} = 
    \bm{E[\bm{w}^{\mathsf{T}}\bm{U}(\bm{\Sigma}^{-1})^{\mathsf{T}}\bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{w}]}
\end{equation*}
$$
Since $\bm{\Sigma}$ in our case is a square diagonal matrix, then $\bm{\Sigma}^{-1} = (\bm{\Sigma}^{-1})^{\mathsf{T}}$. Therefore we could write $\bm{w}^{\mathsf{T}}\bm{U}(\bm{\Sigma}^{-1})^{\mathsf{T}}\bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{w}$ as
$$
\begin{equation*}
    \bm{w}^{\mathsf{T}}\bm{U}\bm{\Sigma}^{-1}\bm{\Sigma}^{-1}\bm{U}^{\mathsf{T}}\bm{w} = \bm{w}^{\mathsf{T}}\bm{U}
    \begin{bmatrix}
     \frac{1}{\bm{\lambda _1^2}} & 0 & 0 & 0 & \cdots & 0 \\
        0 & \frac{1}{\bm{\lambda _2^2}} & 0 & 0 & \cdots & 0 \\
        0 & 0 & \frac{1}{\bm{\lambda _3^2}} & 0 & \cdots & 0 \\
        \vdots & & & \ddots & & \vdots\\
        0 & 0 & 0 & 0 & \cdots & \frac{1}{\bm{\lambda _n^2}}
    \end{bmatrix}
    \bm{U}^{\mathsf{T}}\bm{w}
\end{equation*}
$$
Since the dimension of $\bm{w}$ is $\bm{n}\times 1$, and dimension of $\bm{U}^{\mathsf{T}}$ is $\bm{n} \times \bm{n}$, let's set $\bm{U}^{\mathsf{T}}\bm{w} = \bm{x}$, then $\bm{w}^{\mathsf{T}}\bm{U} = \bm{x}^{\mathsf{T}}$. Therefore 
$$
\begin{equation*}
    \bm{w}^{\mathsf{T}}\bm{U}
    \begin{bmatrix}
     \frac{1}{\bm{\lambda _1^2}} & 0 & 0 & 0 & \cdots & 0 \\
        0 & \frac{1}{\bm{\lambda _2^2}} & 0 & 0 & \cdots & 0 \\
        0 & 0 & \frac{1}{\bm{\lambda _3^2}} & 0 & \cdots & 0 \\
        \vdots & & & \ddots & & \vdots\\
        0 & 0 & 0 & 0 & \cdots & \frac{1}{\bm{\lambda _n^2}}
    \end{bmatrix}
    \bm{U}^{\mathsf{T}}\bm{w} = \bm{x}^{\mathsf{T}}
    \begin{bmatrix}
     \frac{1}{\bm{\lambda _1^2}} & 0 & 0 & 0 & \cdots & 0 \\
        0 & \frac{1}{\bm{\lambda _2^2}} & 0 & 0 & \cdots & 0 \\
        0 & 0 & \frac{1}{\bm{\lambda _3^2}} & 0 & \cdots & 0 \\
        \vdots & & & \ddots & & \vdots\\
        0 & 0 & 0 & 0 & \cdots & \frac{1}{\bm{\lambda _n^2}}
    \end{bmatrix}\bm{x} = \begin{bmatrix}
    \bm{x_1} \\
    \bm{x_2} \\
    \bm{x_3} \\
    \vdots \\
    \bm{x_n}
    \end{bmatrix}^{\mathsf{T}}
     \begin{bmatrix}
     \frac{1}{\bm{\lambda _1^2}} & 0 & 0 & 0 & \cdots & 0 \\
        0 & \frac{1}{\bm{\lambda _2^2}} & 0 & 0 & \cdots & 0 \\
        0 & 0 & \frac{1}{\bm{\lambda _3^2}} & 0 & \cdots & 0 \\
        \vdots & & & \ddots & & \vdots\\
        0 & 0 & 0 & 0 & \cdots & \frac{1}{\bm{\lambda _n^2}}
    \end{bmatrix}
    \begin{bmatrix}
    \bm{x_1} \\
    \bm{x_2} \\
    \bm{x_3} \\
    \vdots \\
    \bm{x_n}
    \end{bmatrix}
\end{equation*}
$$
which can be written as 
$$
\begin{equation*}
\begin{bmatrix}
    \bm{x_1} &
    \bm{x_2} &
    \bm{x_3} &
    \cdots &
    \bm{x_n}
    \end{bmatrix}
\end{equation*}
\begin{bmatrix}
    \frac{\bm{x_1}}{\bm{\lambda _1^2}} \\
    \frac{\bm{x_2}}{\bm{\lambda _2^2}} \\
    \frac{\bm{x_3}}{\bm{\lambda _3^2}} \\
    \vdots \\
    \frac{\bm{x_n}}{\bm{\lambda _n^2}}
    \end{bmatrix} = \frac{\bm{x_1^2}}{\bm{\lambda _1^2}} + \frac{\bm{x_2^2}}{\bm{\lambda _2^2}} + \frac{\bm{x_3^2}}{\bm{\lambda _3^2}} + \bm{\cdots} + \frac{\bm{x_n^2}}{\bm{\lambda _n^2}}
$$
Therefore, the expectation of error can be rewritten as 
$$
\begin{equation*}
    \bm{E\left[\frac{\bm{x_1^2}}{\bm{\lambda _1^2}} + \frac{\bm{x_2^2}}{\bm{\lambda _2^2}} + \frac{\bm{x_3^2}}{\bm{\lambda _3^2}} + \bm{\cdots} + \frac{\bm{x_n^2}}{\bm{\lambda _n^2}}\right]} = 
    \sum_{i = 1}^{n} \frac{\bm{E[\bm{\bm{x_i^2}}]}}{\bm{\lambda _i^2}}
\end{equation*}
$$
Now considering about $\bm{E[\bm{\bm{x_i^2}}]}$, since $\bm{Var[x_i]} = \bm{E[x_i^2]} - \bm{E^2[x_i]}$, then $\bm{E[x_i^2]} = \bm{Var[x_i]} + \bm{E^2[x_i]}$. Because $\bm{x} = \bm{U}^{\mathsf{T}}\bm{w}$, we could write $\bm{x_i} = \bm{u_{1i}\bm{w_1}} + \bm{u_{2i}\bm{w_2}} + \cdots + \bm{u_{ni}\bm{w_n}}$, then
$$
\begin{equation*}
    \bm{E[x_i^2]} = \bm{Var[\bm{u_{1i}\bm{w_1}} + \bm{u_{2i}\bm{w_2}} + \cdots + \bm{u_{ni}\bm{w_n}}]} + \bm{E^2[\bm{u_{1i}\bm{w_1}} + \bm{u_{2i}\bm{w_2}} + \cdots + \bm{u_{ni}\bm{w_n}}]} 
\end{equation*}
$$
which can be rewritten as 
$$
\begin{equation*}
    \bm{E[x_i^2]} = \sum _{j=1}^{n} \bm{Var[u_{ji}w_j]} + \left(\sum _{j=1}^{n}\bm{E[u_{ji}w_j]}\right)^2 = \sum _{j=1}^{n} \bm{u_{ji}^2}\bm{Var[w_j]} + \left(\sum _{j=1}^{n}\bm{u_{ji}}\bm{E[w_j]}\right)^2
\end{equation*}
$$
Since we know that all $\bm{w_i}$ has the same propabality distribution with mean equal to $\bm{\mu}$, and variance equal to $\bm{\sigma}^2$, then we could write the above equation as
$$
\begin{equation*}
   \bm{E[x_i^2]} = \bm{\sigma ^2}\sum _{j=1}^{n} \bm{u_{ji}^2} + \bm{\mu ^2}\left(\sum _{j=1}^{n}\bm{u_{ji}}\right)^2
\end{equation*}
$$
Replace this representation into the equation of expectation of error, we could get
$$
\begin{equation*}
    \bm{E}\left[\| \bm{\hat{x}} - \bm{x_0} \|^2\right] = \sum_{i = 1}^{n} \frac{\bm{\sigma ^2}\sum _{j=1}^{n} \bm{u_{ji}^2} + \bm{\mu ^2}\left(\sum _{j=1}^{n}\bm{u_{ji}}\right)^2}{\bm{\lambda _i^2}}
\end{equation*}
$$
We could change the form of above equation a little bit to have a clarer vision of writing the code for it, we could write it as
$$
\begin{equation*}
    \bm{E}\left[\| \bm{\hat{x}} - \bm{x_0} \|^2\right] = \bm{\sigma ^2}\sum_{i = 1}^{n} \frac{\sum _{j=1}^{n} \bm{u_{ji}^2}}{\bm{\lambda _i^2}} + \bm{\mu ^2}\sum_{i = 1}^{n} \frac{\left(\sum _{j=1}^{n}\bm{u_{ji}}\right)^2}{\bm{\lambda _i^2}} 
\end{equation*}
$$
From this expression, first looking at the first sum term, $\sum _{j=1}^{n} \bm{u_{ji}^2}$ can be calculated by finding the inner product between each column vector of $\bm{U}$ and itself, and then put this value into a new vector, then the first sum term can be calculated as finding the inner product between this vector and the vector that contains all the square inverse of all the elements of original diagonal matrix $\bm{\Sigma}$. The second sum term can be found similarly, but just $\left(\sum _{j=1}^{n}\bm{u_{ji}}\right)^2$ can be calculated by frist adding all the row vectors of $\bm{U}$ together and then finding the element product of this addition vector and itself.

Since we have a requirement that random variables $\bm{w_i}$ will be normalized such that $\| \bm{w} \|_2 = 1$, which means that $\bm{w_i}$ will be normalized to $\frac{\bm{w_i}}{\sum _{k=1}^{n}\bm{w_k^2}}$, then the representation of the error expectation will be modified as 
$$
\begin{equation*}
     \bm{E\left[\|\hat{x} - x_0 \|^2\right]} = \sum _{j=1}^{n} \bm{u_{ji}^2}\bm{Var\left[\frac{\bm{w_j}}{\sum _{k=1}^{n}\bm{w_k^2}}\right]} + \left(\sum _{j=1}^{n}\bm{u_{ji}}\bm{E\left[\frac{w_j}{\sum _{k=1}^{n}\bm{w_k^2}}\right]}\right)^2 = \bm{\frac{1}{\left(\sum _{k=1}^{n}\bm{w_k^2}\right)^2}}\left(\sum _{j=1}^{n} \bm{u_{ji}^2}\bm{Var[w_j]} + \left(\sum _{j=1}^{n}\bm{u_{ji}}\bm{E[w_j]}\right)^2\right)
\end{equation*}
$$
Therefore, the final representation of error expecation under this constraint would become
$$
\begin{equation*}
    \bm{E}\left[\| \bm{\hat{x}} - \bm{x_0} \|^2\right] = \bm{\frac{1}{\left(\sum _{k=1}^{n}\bm{w_k^2}\right)^2}}\left(\bm{\sigma ^2}\sum_{i = 1}^{n} \frac{\sum _{j=1}^{n} \bm{u_{ji}^2}}{\bm{\lambda _i^2}} + \bm{\mu ^2}\sum_{i = 1}^{n} \frac{\left(\sum _{j=1}^{n}\bm{u_{ji}}\right)^2}{\bm{\lambda _i^2}}\right)
\end{equation*}
$$
Since in our case, $\bm{\|w\|^2 = 1}$, then $\sum _{k=1}^{n}\bm{w_k^2} = 1$, then 
$$
\begin{equation*}
    \bm{E}\left[\| \bm{\hat{x}} - \bm{x_0} \|^2\right] = \bm{\sigma ^2}\sum_{i = 1}^{n} \frac{\sum _{j=1}^{n} \bm{u_{ji}^2}}{\bm{\lambda _i^2}} + \bm{\mu ^2}\sum_{i = 1}^{n} \frac{\left(\sum _{j=1}^{n}\bm{u_{ji}}\right)^2}{\bm{\lambda _i^2}}
\end{equation*}
$$
In our case, if we assume it's a normal distribution, then $\bm{\mu} = 0$ and $\bm{\bm{\sigma}} = 1$, then 
$$
\begin{equation*}
    \bm{E}\left[\| \bm{\hat{x}} - \bm{x_0} \|^2\right] = \sum_{i = 1}^{n} \frac{\sum _{j=1}^{n} \bm{u_{ji}^2}}{\bm{\lambda _i^2}}
\end{equation*}
$$

```julia
function Find_error_expectation(mu, sigma, A)
    #find the SVD of A to get Σ and U
    F = svd(A)
    U = F.U
    Σ = F.S
    #find the inverse of λ_i^2
    λ_inv = 1 ./ (Σ .* Σ)
    n = length(A[1, :])
    first_sum_num = zeros(n)
    #find the first sum numerator vector
    for i in 1:n
        first_sum_num[i] = U[:, i]'*U[:, i]
    end
    first_sum = first_sum_num'*λ_inv
    #find the second sum numerator vector
    second_sum_num = U[1, :]
    for i in 2:n
        second_sum_num = second_sum_num + U[i, :]
    end
    second_sum_num = second_sum_num .* second_sum_num
    second_sum = second_sum_num'*λ_inv
    #find the error error expectation
    error_expectation = sigma*sigma*first_sum + mu*mu*second_sum
    return error_expectation
end

e = Find_error_expectation(0, 1, CW1_DATA.Q5_2_data_A)
CW1_DATA.Q5_2_c_ans_err = e
```

## 5.3. LS Recovery from Convolution

We could assume the cyclic convolution and subsampling operation has corresponded operator, then it can be represented by 
$$
\begin{equation*}
    \frac{1}{2}\|\bm{y}-\bm{P_{\Omega}Hx}\|^2 + \frac{\bm{\alpha}}{2}\|\bm{x}\|^2
\end{equation*}
$$
Let $\bm{P_{\Omega}H} = \bm{\hat{H}}$, we could get that 
$$
\begin{equation*}
    \frac{1}{2}\left(\bm{y} - \bm{\hat{H}x}\right)^{\mathsf{T}}\left(\bm{y} - \bm{\hat{H}x}\right) + \frac{\bm{\alpha}}{2}\bm{x}^{\mathsf{T}}\bm{x} = \frac{1}{2}\left(\bm{y}^{T} - \bm{x}^{\mathsf{T}}\bm{\hat{H}}^{\mathsf{T}}\right)\left(\bm{y - \hat{H}x}\right) + \frac{\bm{\alpha}}{2}\bm{x}^{\mathsf{T}}\bm{x} = \frac{1}{2}\left(\bm{y}^{\mathsf{T}}\bm{y} - \bm{x}^{\mathsf{T}}\bm{\hat{H}^{\mathsf{T}}}\bm{y} - \bm{y^{\mathsf{T}}}\bm{\hat{H}}\bm{x} + \bm{x^{\mathsf{T}}}\bm{\hat{H}^{\mathsf{T}}}\bm{\hat{H}x}\right) + \frac{\bm{\alpha}}{2}\bm{x^{\mathsf{T}}x}
\end{equation*}
$$
Which can be reformed as 
$$
\begin{equation*}
    \frac{1}{2}\bm{x^{\mathsf{T}}}\left(\bm{\hat{H}^{\mathsf{T}}\hat{H}} + \bm{\alpha}\right)\bm{x} - \bm{y^{\mathsf{T}}\hat{H}x} + \frac{1}{2}\bm{y^{\mathsf{T}}y}
\end{equation*}
$$
It can be seen that in this case
$$
\begin{equation*}
\begin{split}
    \bm{A} &= \bm{\hat{H}^{\mathsf{T}}\hat{H}} + \bm{\alpha}\bm{I} \\
    \bm{b} &= \left(-\bm{y^{\mathsf{T}}H}\right)^{\mathsf{T}}
\end{split}
\end{equation*}
$$


### 1.(a)

```julia
function Find_CyclicConvolution_Matrix(h)
    n = length(h)
    h_new = vcat(h[2:n], h)
    C = zeros(n, n)
    for i in 1:n
        C[i, :] = reverse(h_new[i:i+n-1])
    end
    return C
end

function Find_p_omega(Omega, n)
    m = length(Omega)
    O = zeros(m, n)
    for i in 1:m
        O[i, Omega[i]] = 1
    end
    return O
end

function Find_A_b(alpha, h, n, Omega, y)
    H = Find_CyclicConvolution_Matrix(h)
    P_omega = Find_p_omega(Omega, n)
    P_H = P_omega*H
    A = transpose(P_H)*P_H + alpha*I(n)
    b = transpose(-transpose(y)*P_H)
    return A, b     
end

A, b = Find_A_b(CW1_DATA.Q5_3_data_alpha, CW1_DATA.Q5_3_data_h, CW1_DATA.Q5_3_data_n, CW1_DATA.Q5_3_data_Omega, CW1_DATA.Q5_3_data_y)
CW1_DATA.Q5_3_1_a_ans_A = A
CW1_DATA.Q5_3_1_a_ans_b = b
```

### 2.(a)

Based on the previous derivation, we could see that 
$$
\begin{equation*}
    \bm{u^{\mathsf{T}}Av} = \bm{u^{\mathsf{T}}\left(\hat{H}^{\mathsf{T}}\hat{H} + \alpha \right)x} = \bm{u^{\mathsf{T}}\hat{H}^{\mathsf{T}}\hat{H}v} + \bm{\alpha u^{\mathsf{T}}v} = \left(\bm{\hat{H}u}\right)^{\mathsf{T}}\bm{\hat{H}v} + \bm{\alpha u^{\mathsf{T}}v}
\end{equation*}
$$
which can be written as 
$$
\begin{equation*}
    \left(\bm{P_{\Omega}(h\circledast _n u)}\right)^{\mathsf{T}}\bm{P_{\Omega}(h\circledast _n v)} + \bm{\alpha u^{\mathsf{T}}v}
\end{equation*}
$$

```julia
function Find_Cyclic_Convulotion_withoutMatrix(x, h)
    n = length(x)
    k = DSP.conv(h, x)
    y = k[1:n] + vcat(k[(n+1):length(k)], 0)
    return y 
end

function Find_uAv_bu_withoutmatrix(u, h, v, Omega, alpha, y)
    Hu = Find_Cyclic_Convulotion_withoutMatrix(u, h)
    Hv = Find_Cyclic_Convulotion_withoutMatrix(v, h)
    m = length(Omega)
    OHu = zeros(m)
    OHv = zeros(m)
    for i in 1:m
        OHu[i] = Hu[Omega[i]]
        OHv[i] = Hv[Omega[i]]
    end
    uAv = sum(OHu .* OHv) + alpha*sum(u .* v)
    bu = -sum(y .* OHu)
    return uAv, bu
end

uAv, bu = Find_uAv_bu_withoutmatrix(CW1_DATA.Q5_3_data_u, CW1_DATA.Q5_3_data_h, CW1_DATA.Q5_3_data_v, CW1_DATA.Q5_3_data_Omega, CW1_DATA.Q5_3_data_alpha, CW1_DATA.Q5_3_data_y)
CW1_DATA.Q5_3_2_a_ans_uAv = uAv
CW1_DATA.Q5_3_2_a_ans_bu = bu

```

### 2.(b)

```julia
function evaluate_adjoint_subsample(n,m,omega,y)
    z = zeros(n)  
    for i in 1:m
        index = omega[i]
        z[index] = y[i]
    end
    return z
end

function Find_Cyclic_Convolution_Adjoint_withoutMatrix(y, h)
    n = length(h)
    h_new = vcat(h[1], reverse(h[2:n]))
    k = DSP.conv(h_new, y)
    z = k[1:n] + vcat(k[(n+1):length(k)], 0)
    return z  
end

function Find_Au_withoutmatrix(u, h, Omega, alpha)
    Hu = Find_Cyclic_Convulotion_withoutMatrix(u, h)
    m = length(Omega)
    OHu = zeros(m)
    # calculate P_OmegaH operator
    for i in 1:m
        OHu[i] = Hu[Omega[i]]
    end
    # calculate P_Omega^TP_OmegaH operator
    n = length(u)
    O_TOHu = evaluate_adjoint_subsample(n, m, Omega, OHu)
    # calculate H^TP_Omega^TP_OmegaH operator
    H_TO_TOHu = Find_Cyclic_Convolution_Adjoint_withoutMatrix(O_TOHu, h)
    # calculate Au = H^TP_Omega^TP_OmegaHu + alpha*u
    Au = H_TO_TOHu + alpha*u
    return Au
end

function Find_bu_withoutmatrix(u, h, Omega, alpha, y)
    Hu = Find_Cyclic_Convulotion_withoutMatrix(u, h)
    m = length(Omega)
    OHu = zeros(m)
    for i in 1:m
        OHu = Hu[Omega[i]]
    end
    bu = -sum(y .* OHu)
    return bu
end

function Find_LiearOperator(h, Omega, alpha, y)
    n = length(h)
    A = LinearOperator(Float64, n, n, false, false, (Au, u) -> Au .= Find_Au_withoutmatrix(u, h, Omega, alpha))
    b_T = LinearOperator(Float64, n, 1, false, false, (bu, u) -> bu .= Find_bu_withoutmatrix(u, h, Omega, alpha, y))
    return A, b_T
end

function Find_Conjugate_Gradient(h, Omega, alpha, y)
    A, b = Find_LiearOperator(h, Omega, alpha, y)
    n = length(h)
    A_2, b_2 = Find_A_b(alpha, h, n, Omega, y)
    return cg(A, -b_2)
end

x_out = Find_Conjugate_Gradient(CW1_DATA.Q5_3_data_h, CW1_DATA.Q5_3_data_Omega, CW1_DATA.Q5_3_data_alpha, CW1_DATA.Q5_3_data_y)
CW1_DATA.Q5_3_2_b_ans_x = x_out
```

```julia
function Find_POmega_withoutmatrix(Omega, X)
    m, n = size(Omega)
    X_new = []
    for j in 1:n
        for i in 1:m
            X_new = (Omega[i, j] == 1) ? vcat(X_new, X[i, j]) : X_new
        end
    end
    return X_new
end

function Find_MRI_operation_2(Omega, X)
    m, n = size(X)
    ##### Find O1 #######
    # First compute the 1D DFT for column vectors of X
    O_1_pre = zeros(ComplexF64, m, n)
    for i in 1:n
        O_1_pre[:, i] = FFTW.fft(X[:, i])
    end
    # Then compute the 1D DFT for row vectors of O_1_pre
    O_1 = zeros(ComplexF64, m, n)
    for i in 1:n
        O_1[i, :] = FFTW.fft(O_1_pre[i, :])
    end
    ##### Find O2 #######
    # First compute the 1D fftshift for row vectors of O1
    O_2_pre = zeros(ComplexF64, m, n)
    for i in 1:m
        O_2_pre[i, :] = FFTW.fftshift(O_1[i, :])
    end
    # Then compute the 1D fftshift for column vectors of O_2_pre
    O_2 = zeros(ComplexF64, m, n)
    for i in 1:n
        O_2[:, i] = FFTW.fftshift(O_2_pre[:, i])
    end
    ##### Find O3 #######
    O_3 = Find_POmega_withoutmatrix(Omega, O_2)
    return O_3
end

y = Find_MRI_operation_2(CW1_DATA.Q5_4_data_Omega, CW1_DATA.Q5_4_data_X)
CW1_DATA.Q5_4_1_a_ans_y = y
```

### 1.(b)

The basic procedure to find $\bm{A}$ is same as 5.3, but since DFT is a complex operator, we consider $\|\bm{y - P_{\Omega}\circ S \circ F_{2D}(x)} \|^2$ as the Frobenius norm, which is 
$$
\begin{equation*}
    \|\bm{y - P_{\Omega}\circ S \circ F_{2D}(x)} \|_F^2 = \sum _{i=1}^{n}\left|\left(\bm{y - P_{\Omega}\circ S \circ F_{2D}(x)}\right)_i\right|^2
\end{equation*}
$$
Now assume that $P_{\Omega}\circ S \circ F_{2D}(x)$ can be equivalent to $\bm{\hat{F}x}$, then the Frobenius norm can rewrite as 
$$
\begin{equation*}
    \sum _{i=1}^{n}\left|\left(\bm{y - \hat{F}x}\right)_i\right|^2 = \sum _{i=1}^{n}\left(\bm{y - \hat{F}x}\right)_i^{\ast}\left(\bm{y - \hat{F}x}\right)_i = \left(\left(\bm{y - \hat{F}x}\right)^{\ast}\right)^{\mathsf{T}}\left(\bm{y - \hat{F}x}\right)
\end{equation*}
$$
which can be written as 
$$
\begin{equation*}
    \left(\bm{y^{\ast} - \hat{F}^{\ast}x^{\ast}}\right)^{\mathsf{T}}\left(\bm{y - \hat{F}x}\right)
\end{equation*}
$$
Since $\bm{x}$ is real, then we could write that as
$$
\begin{equation*}
    \left(\bm{y^{\ast} - \hat{F}^{\ast}x}\right)^{\mathsf{T}}\left(\bm{y - \hat{F}x}\right) = \left(\bm{(y^{\ast})^{\mathsf{T}}} - \bm{x^{\mathsf{T}}\bm{\left(\hat{F}^{\ast}\right)^{\mathsf{T}}}}\right)\left(\bm{y - \hat{F}x}\right) = \bm{x^{\mathsf{T}}\hat{F}^{\ast \mathsf{T}}\hat{F}x} - \bm{y^{\mathsf{T}}\hat{F}^{\ast}x} - \bm{y^{\ast \mathsf{T}}\hat{F}x} + \bm{y^{\ast \mathsf{T}}y}
\end{equation*}
$$
Therefore, it can be seen that 
$$
\begin{equation*}
\begin{split}
    \bm{u^{\mathsf{T}}Av} &= \bm{\left(\left(P_{\Omega}\circ S \circ F_{2D}(u)\right)^{\ast}\right)^{\mathsf{T}}P_{\Omega}\circ S \circ F_{2D}(v)} + \bm{\alpha u^{\mathsf{T}}v} \\
    \bm{b^{\mathsf{T}}z} &= - \bm{y^{\mathsf{T}}\left(P_{\Omega}\circ S \circ F_{2D}(z)\right)^{\ast}} - \bm{y^{\ast \mathsf{T}}P_{\Omega}\circ S \circ F_{2D}(z)}
\end{split}
\end{equation*}
$$

```julia
function Find_uAv_MRI(u, v, Omega, alpha)
    m, n = size(Omega)
    V = zeros(ComplexF64, m, n)
    U = zeros(ComplexF64, m, n)
    for i in 1:n
        V[:, i] = v[(i-1)*m+1 : (i-1)*m+m]
        U[:, i] = u[(i-1)*m+1 : (i-1)*m+m]
    end
    O_3_U = Find_MRI_operation_2(Omega, U)
    O_3_V = Find_MRI_operation_2(Omega, V)
    uAv = transpose(conj(O_3_U))*O_3_V + alpha*u'*v
    return uAv
end

function Find_bz_MRI(z, y, Omega)
    m, n = size(Omega)
    Z = zeros(m, n)
    for i in 1:n
        Z[1:end, i] = z[(i-1)*m+1 : (i-1)*m+m]
    end
    O_3_Z = Find_MRI_operation_2(Omega, Z)
    bv = 0.5*(-transpose(y)*conj(O_3_Z) - transpose(conj(y))*O_3_Z)
    return bv
end

uAv = Find_uAv_MRI(CW1_DATA.Q5_4_data_u, CW1_DATA.Q5_4_data_v, CW1_DATA.Q5_4_data_Omega, CW1_DATA.Q5_4_data_alpha)
bz = Find_bz_MRI(CW1_DATA.Q5_4_data_z, CW1_DATA.Q5_4_1_a_ans_y, CW1_DATA.Q5_4_data_Omega)
CW1_DATA.Q5_4_1_b_ans_uAv = real(uAv)
CW1_DATA.Q5_4_1_b_ans_bz = real(bz)
```

In this question, we need to find the function to find $\bm{Au}$ to generate the linearoperator, and we also need to find the matrix representation of $\bm{b}$ so that we could use the cg function. First considering about finding the $\bm{Au}$, which is equal to firsting doing MRI operation for $\bm{u}$, and then do the adjoint MRI operation to it.

To find $\bm{b}$, we have to find the matrix representation of it for cg function, however it would be very time consuming to the matrix representation of $\bm{P_{\Omega}}$, so it's important to find a way to directly calculate the $\bm{b}$ without generating the matrix of $\bm{P_{\Omega}}$

If writing the MRI operations into operator form, we could see that 
$$
\begin{equation*}
    \bm{b^{\mathsf{T}}} = \bm{\frac{1}{2}}\left(- \bm{y^{\mathsf{T}}\left(P_{\Omega}SF_{2D}\right)^{\ast}} - \bm{y^{\ast \mathsf{T}}P_{\Omega}SF_{2D}}\right)

Therefore
$$
\begin{equation*}
    \bm{b} = \left(\bm{\frac{1}{2}}\left(- \bm{y^{\mathsf{T}}\left(P_{\Omega}SF_{2D}\right)^{\ast}} - \bm{y^{\ast \mathsf{T}}P_{\Omega}SF_{2D}}\right)\right)^{\mathsf{T}} = \bm{\frac{1}{2}}\left(- \bm{\left(P_{\Omega}SF_{2D}\right)^{\ast\mathsf{T}}y} - \bm{\left(P_{\Omega}SF_{2D}\right)^{\mathsf{T}}y^{\ast}}\right)
\end{equation*}
$$
Reform the above equation we could get 
$$
\begin{equation*}
    \bm{b} = \bm{\frac{1}{2}}\left(- \bm{\left(P_{\Omega}SF_{2D}\right)^{\ast\mathsf{T}}y} - \bm{\left(\left(P_{\Omega}SF_{2D}\right)^{\ast\mathsf{T}}y\right)^{\ast}}\right)
\end{equation*}
$$
In this form, we could use the matrix free MRI adjoint operation function we've already designed to get $\bm{b}$.

```julia
function Adjoint_Subsampling(Omega, y)
    m, n = size(Omega)
    Z = zeros(ComplexF64, m, n)
    index = 1
    for j in 1:n
        for i in 1:m
            Z[i, j] = (Omega[i, j] == 0) ? 0 : y[index]
            index = (Omega[i, j] == 0) ? index : index+1
        end
    end
    return Z
end

function Find_Au_MRI(u, Omega, alpha)
    m, n = size(Omega)
    U = zeros(ComplexF64, m, n)
    for i in 1:n
        U[:, i] = u[(i-1)*m+1 : (i-1)*m+m]
    end
    O_3_U = Find_MRI_operation_2(Omega, U)
    O1, O2, O3 = Find_Adjoint_MRI_withoutmatrix(Omega, O_3_U)
    # vectorize O3
    l, h = size(O3)
    out = zeros(ComplexF64, l*h)
    for i in 1:h
        out[(i-1)*l+1 : i*l] = O3[1:end, i]
    end
    out = out + alpha*u
    return out
end

function Find_LiearOperator_MRI(Omega, alpha)
    m, n = size(Omega)
    l = m*n
    A = LinearOperator(ComplexF64, l, l, false, false, (Au, u) -> Au .= Find_Au_MRI(u, Omega, alpha))
    return A
end

function Find_b_MRI(Omega, y)
    O1_1, O2_1, O3_1 = Find_Adjoint_MRI_withoutmatrix(Omega, y)
    l, h = size(O3_1)
    first_term = zeros(ComplexF64, l*h)
    for i in 1:h
        first_term[(i-1)*l+1 : i*l] = O3_1[1:end, i]
    end
    O1_2, O2_2, O3_2 = Find_Adjoint_MRI_withoutmatrix(Omega, y)
    m, n = size(O3_2)
    second_term = zeros(ComplexF64, m*n)
    for i in 1:n
        second_term[(i-1)*l+1 : i*l] = O3_2[1:end, i]
    end
    b = 0.5*(-first_term-conj(second_term))
    return b
end


function Find_Conjugate_Gradient_MRI(Omega, alpha, y)
    A = Find_LiearOperator_MRI(Omega, alpha)
    b = Find_b_MRI(Omega, y)
    out = cg(A, -b)
    m, n = size(Omega)
    X_hat = zeros(ComplexF64, m, n)
    for i in 1:n
        X_hat[1:end, i] = out[(i-1)*m+1 : i*m]
    end
    return X_hat
end

X_hat = Find_Conjugate_Gradient_MRI(CW1_DATA.Q5_4_data_Omega, CW1_DATA.Q5_4_data_alpha, CW1_DATA.Q5_4_1_a_ans_y)
CW1_DATA.Q5_4_1_c_ans_X_hat = X_hat
```
